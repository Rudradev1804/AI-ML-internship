```python
# Q-3 Write a function add_two_numbers() that takes two numbers as input (using input function) and prints their sum.

def add_two_numbers():
    a = eval(input("Enter first number:"))
    b = eval(input("Enter second number:"))
    print(a+b)

add_two_numbers()
```

```python
# Q-5 Write a function print_details() that prints your name, course, and city.

def print_details(name, course, city):
    name = name
    course = course
    city = city

    return name, course, city

print("Rudra", "B.Tech", "Ahmedabad")

x,y,z = print_details("Rudra", "B.Tech", "Ahmedabad")
print("name", x)
print("course", y)
print("city", z)
```

```python
# Q-8 Write a function area_of_circle(radius) that returns the area of a circle.

def area_of_circle(radius):
    pi = 3.14
    area = pi *(radius**2)
    return area

area_of_circle(7)
```

```python
# Q-9 Write a function average_of_three(a, b, c) that returns the average of three numbers.

def average_of_three(a, b, c):
    avg = (a+b+c)/3
    return avg

average_of_three(43, 21, 17)
```

```python
# Q-10 Write a function is_even(num) that returns True if number is even, otherwise False.

def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False
    
is_even(44)
```

```python
# Q-11 Write a function power(base, exponent=2) that returns base raised to exponent.

def power(base, exceponent=2):
    return base ** exceponent

power(3, 5)
```

```python
# Q-13 Write a function student_info(name, age, course) and call it using keyword arguments (in any order).

def student_info(**details):
    for key, value in details.items():
        print(f"{key}: {value}")

student_info(name="Rudra", age=21, course="B.Tech")
```

```python
# Q-16 Write a function find_max(*nums) that returns the maximum number among all arguments.

def find_max(*nums):
    return max(nums)
find_max(4, 7, 3, 8, 5)
```

```python
# Q-17 Write a function show_details(**info) that prints key-value pairs of employee details.

def show_details(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

show_details(name="Rudra", age=21, city="Ahmedabad")
```

```python
# Q-19 Write a function calculate_marks(english, maths, science) that returns total marks and average marks.

def calculate_marks(english, maths, science):
    total_marks = english + maths + science
    average = total_marks/3
    return total_marks, average

total_marks, average = calculate_marks(85, 90, 78)
print("Total Marks:", total_marks)
print("Average Marks:", average)
```
