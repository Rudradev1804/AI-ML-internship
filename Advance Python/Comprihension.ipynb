```python
# Q-1 Create a list of all numbers from 1 to 10 using list comprehension.

numbers = [i for i in range(1, 11)]
print(numbers)
```

```python
# Q-2 Create a list of squares of numbers from 1 to 10.

squares = [i**2 for i in range(1, 11)]
print(squares)
```

```python
# Q-3 Create a list of even numbers from 1 to 20.

even = [x for x in range(1, 11) if x%2==0]
print(even)
```

```python
# Q-4 Create a list of cubes of odd numbers from 1 to 10.

cubes = [y**3 for y in range(1, 11)]
print(cubes)
```

```python
# Q-5 Convert all words in a list to uppercase using comprehension.

uppercase = [word.upper() for word in ['Hello','rudra','welcome to the', 'python']]
print(uppercase)
```

```python
# Q-6 Create a dictionary of numbers and their cubes (1–5).

cubes = {x:x**3 for x in range(1, 6)}
print(cubes)
```

```python
# Q-7 Create a dictionary of numbers from 1–10 with even numbers only.

even_numbers = {i:i for i in range(11) if i%2==0}
print(even_numbers)
```

```python
# Q-8 Create a list of tuples (x, y) for x in [1,2,3] and y in [4,5,6].

tuples = [(x, y) for x in [1,2,3] for y in [4,5,6]]
print(tuples)
```

```python
# Q-9 Use a generator comprehension to print squares of 1–10.

gen = (x*2 for x in range(1, 11))

for value in gen:
    print(value)
```

```python
# Q-10 Create a set of unique vowels from a given string using comprehension.

strings = "Hello Rudra welcome to the world of Pyhton Programming"
vowels = (char for char in strings if char in 'aeiouAEIOU')
unique_vowels = set(vowels)
print(unique_vowels)
```

```python
# Q-11 Create a list of factorials for numbers from 1 to 5 using comprehension.

import math
fact = [math.factorial(i) for i in range(1, 6)]
print(fact)
```
