```python
# Dataset 1 â€” Students Marks

# marks = np.array([35, 50, 65, 80, 40, 90, 55, 70])

# Q11: Find indexes of students who passed (marks â‰¥ 50).

# Q12: Replace all marks below 50 with 0 using np.where().

# Q13: Find where a new score 75 should be inserted to keep marks sorted.

# Q14: Get all scores greater than average marks.
import numpy as np

marks = np.array([35, 50, 65, 80, 40, 90, 55, 70])
avg = np.where(marks > 50)
print(avg)

rpl = np.where(marks<50, 0, marks)
print(rpl)

new = np.searchsorted(marks, 75)
print(new)

great = np.mean(marks)
greater = marks[marks>great]
print(great)
print(greater)
```

```python
# ðŸ§© Dataset 2 â€” Product Prices

# prices = np.array([1200, 800, 1500, 2500, 1800, 900, 3000])

# Q15: Find positions of products cheaper than â‚¹1000.

# Q16: Replace prices above â‚¹2000 with 'Expensive', others with 'Affordable'.

# Q17: Using np.searchsorted(), find where â‚¹1600 should be inserted to maintain order.

# Q18: Get all product prices between â‚¹1000 and â‚¹2000 using boolean indexing.

prices = np.array([1200, 800, 1500, 2500, 1800, 900, 3000])
cheap = np.where(prices>1000)
print(cheap)

pla = np.where(prices>2000, 'Expensive', 'Affordable')
print(pla)

price = np.searchsorted(prices, 1600)
print(price)

beet = np.where((prices>1000) & (prices<2000))
print(beet)
```
