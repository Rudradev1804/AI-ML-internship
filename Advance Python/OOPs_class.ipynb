```python
# Q-5: Create a Circle class with an attribute radius. Write two methods: area() and parameter(). Create an object and call both methods.

class Circle:
    def __init__(self, radius):
        self.radius =radius

    def area(self):
        return 3.14 * self.radius * self.radius
    
    def perimeter(self):
        return 2 * 3.14 * self.radius
    
c1 = Circle (7)
print(f"Area of Circle : {c1.area()}")
print(f"Perimeter of Circle : {c1.perimeter()}")
```

```python
# Q-6: Create a Laptop class with attributes brand,ram and price.Write a method apply_discount(self,percent) that updates the price after applying the discount.

class Laptop:
    def __init__(self, brand, ram, price):
        self.brand = brand
        self.ram = ram
        self.price = price

    def apply_discount(self, percent):
        discount_amount = (percent/100) * self.price
        self.price -= discount_amount
        return self.price
    
l1 = Laptop("HP", "32GB", 132000)
print(f"Price after discount : {l1.apply_discount(10)}")
```

```python
# Q-7: Create a Bank Account class with:

# ->Attributes: account_holder and balance.

# ->Methods: deposite(amount)->increase balance, withdraw(amount)->decrease balance, get_balance()->returns current balance.

class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposite(self, amount):
        self.balance += amount
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient balance"
        self.balance -= amount
        return self.balance
    
    def get_balance(self):
        return self.balance
    
b1 = BankAccount("Rudra", 5000)
print(f"Balance after deposite : {b1.deposite(6000)}")
print(f"Balance after withdraw : {b1.withdraw(2500)}")
print(f"Current Balance : {b1.get_balance()}")
```

```python
# Q-9: Create a shoppingCart class with:

# ==>Attribute: items list (initialized as empty in the constructor).

# ==>Method: add_item(item)->adds item to the list.

# ==>Method: remove_item(item)->removes item from list.

# ==>Method: show_items()-> prints all items in the cart.

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_items(self, item):
        self.items.append(item)

    def remove_items(self, item):
        if item in self.items:
            self.items.remove(item)
        else:
            return "It is not in the cart"
        
    def show_items(self):
        for item in range(len(self.items)):
            print(self.items[item])

cart = ShoppingCart()
cart.add_items("Laptop")
cart.add_items("Smartphone")
cart.add_items("Tablet")
cart.add_items("Headphones")
cart.show_items()
cart.remove_items("Laptop")
```

```python
# Q-3: Create an Employee Class with a private variable __salary.

# -->Write a method increase_salary(percent) that increase the salary based on percentage.

# -->Add a getter method to return updated salary.
class Employee:
    def __init__(self, salary):
        self.__salary = salary

    def increase_salary(self, percent):
        increase_amount = (percent/100)* self.__salary
        self.__salary += increase_amount
        return self.__salary
    def get_salary(self):
        return self.__salary
    
e1 = Employee(50000)
print(f"Updated Salary : {e1.increase_salary(10)}")
```

```python
# Q-4: Create a Mobile class with a private variable __battery.

# -->Write a method charge(percent) that increases the battery level, but it should never go above 100%.

# -->Write another method use(percent) that decreases the battery level, but never goes below 0.

class Mobile:
    def __init__(self, battery):
        self.__battery = battery

    def charge(self, percent):
        self.__battery += percent
        if self.__battery > 100:
            self.__battery = 100
        return self.__battery
    
    def use(self, percent):
        self.__battery -= percent
        if self.__battery < 0:
            self.__battery = 0
        return self.__battery
    
m1 = Mobile(50)
print(f"Battery after charging : {m1.charge(60)}")
print(f"Battery after use : {m1.use(30)}")
```

```python
# Q-6: Create a ShoppingCart class with a private list __items.

# -->Write a method add_item(item) to add item to cart.

# -->Write a method remove_item(item) to remove item safely(check if it exists).

# -->Write a getter method get_items() to return all cart items.

class ShoppingCart:
    def __init__(self):
        self.__items = []

    def add_items(self, items):
        self.__items.append(items)

    def remove_items(self, items):
        if items in self.__items:
            self.__items.remove(items)
        else:
            return "Items not found in the cart"
        
    def get_items(self):
        return self.__items
    
cart = ShoppingCart()
cart.add_items("Camera")
cart.add_items("Laptop")
cart.add_items("Smartphone")
cart.add_items("Mic")
cart.add_items("Monitors")
cart.remove_items("Camera")
print(cart.get_items())
```
