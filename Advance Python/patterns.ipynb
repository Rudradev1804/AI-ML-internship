```python
alpha,digit=ord("a"),1
for i in range(5):
    for j in range(4-i):
        print(' ',end="")
    for k in range(i+1):
        if i%2==0:
            print(digit,end=" ")
            digit+=1
        else:
            print(chr(alpha),end=" ")
            alpha+=1
    print()
alpha,digit=ord("f"),4
for i in range(4):
    for j in range(i+1):
        print(' ',end="")
    for k in range(4-i):
        if i%2==0:
            print(chr(alpha),end=" ")
            alpha-=1
        else:
            print(digit,end=" ")
            digit-=1
    print()
```
