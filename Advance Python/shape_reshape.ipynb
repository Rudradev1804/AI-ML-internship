```python
# Q-1: Create a 1D NumPy array of size 10.

# ->Find its shape,ndim and size.

import numpy as np

arr = np.arange(10)
print("Array:", arr)
print("Shape:", arr.shape)
print("Dimentions:", arr.ndim)
print("Size:", arr.size)
```

```python
# Q-2: Make a 2D array of shape(3,4) with values from 0 to 11.

# ->Print its shape,ndim and size.

arr_2d = np.arange(12).reshape(3, 4)
print(arr_2d)
print("Shape", arr_2d.shape)
print("Dimentions:", arr_2d.ndim)
print("Size:", arr_2d.size)
```

```python
# Q-3: Convert a 1D array of size 12 into:

# ->shape(3,4)

# ->shape(2,2,3)

# ->shape(6,2)

arr_1d = np.arange(12).reshape(3, 4)
print("Shape:", arr_1d.shape)
print(arr_1d)

arr_2d = arr_1d.reshape(2, 2, 3)
print("Shape:", arr_2d.shape)
print(arr_2d)

arr_3d = arr_1d.reshape(6, 2)
print("Shape:", arr_3d.shape)
print(arr_3d)
```

```python
# Q-4: Use -1 in reshape to automatically calculate the missing dimension. Example reshape(3,4) into (2,-1)

a1d = np.arange(12).reshape(3, 4)
print(a1d)
print(a1d.reshape(2, -1))
```

```python
# Q-5: Create a 2D array of shape (3,3) and:

# ->Flatten it into 1D using flatten()

# ->Convert it into 1D using ravel() and change one element.

# ->Check whether original array changes or not.

arr_2d = np.arange(9).reshape(3, 3)
print("Original array:", arr_2d)
arr_f = arr_2d.flatten()
print("Flattened:", arr_f)

arr_r = arr_2d.ravel()
print("Raveled:", arr_r)

arr_r[0] = 99
print("Modified raveled:", arr_r)
print(arr_2d)
```

```python
# Q-6 Take a 1D array [1,2,3,4]

# ->Convert it into shape(1,4) using np.newaxis

# ->Convert it into shape(4,1) using np.expand_dims

arr = np.array([1, 2, 3, 4])
arr_row = arr[np.newaxis, :]
arr_col = (np.expand_dims(arr, axis=1).shape)
arr_co = (np.expand_dims(arr, axis=0).shape)
print("Original array:", arr)
print("Row vector:", arr_row)
print("Row:", arr_co)
print("Column:", arr_col)
```

```python
# Q-7: Create an array of shape(1,5,1) and remove all extra dimensions using squeeze().

arr = np.arange(5).reshape(1, 5, 1)
arr_squeezed = arr.squeeze()
print("Original array:", arr)
print("Squeeze:", arr_squeezed)
```

```python
# -8: Make a 2D array of shape(2,3) and print its transpose.

# ->Verify shape before and after transpose.

arr_2d = np.arange(6).reshape(2, 3)
print("Original:", arr_2d)
print("Shape:", arr_2d.shape)

arr_trans = arr_2d.T
print("Transpose:", arr_trans)
print("Shape of transpose:", arr_trans.shape)
```

```python
# Q-9: Create a 3D array of shape (2,3,4) and use transpose() to swap axes.

# ->Try arr.transpose(1,0,2) and check new shape.

arr_3d = np.arange(24).reshape(2, 3, 4)
print("Original array:", arr_3d)

arr_transposed = arr_3d.transpose(1, 0, 2)
print("Transposed:", arr_transposed)
print("New shape:", arr_transposed.shape)
```

```python
# Q-13:An RGB image has shape(64,64,3)->(height,width,channels).

# (a)Reshape it into (4096,3) so each pixels RGB is a row.

# (b) Suppose you have 100 images->shape(100,64,64,3).

# ->Reshape into (100,12288) for feeding into a neural network.

# ->Transpose a single image into (3,64,64) (channel-first format for PyTorch).

image = np.random.randint(0, 256, (64, 64, 3), dtype=np.uint8)
image_reshaped = image.reshape(4096, 3)
print(image)
print(image.ndim)
print("Reshaped image:", image_reshaped.shape)

image = np.random.randint(0, 256, (100, 64, 64, 3), dtype=np.uint8)
images_reshaped = image.reshape(100, 12288)
print(image.ndim)
print("Reshped images:", images_reshaped.shape)

image_transpose = image[0].transpose(2, 0, 1)
print("Transposed image:", image_transpose.shape)
print(image_transpose)
```

```python
# Q-14:A dataset has shape(10000,20) (10,000 samples,20 features).

# (a)Reshape it into (100,100,20)->100 batches,each with 100 samples.

# (b)Select one batch(100,20) and transpose it into (20,100).

data = np.random.rand(10000, 20)
data_batches = data.reshape(100, 100, 20)
print("Reshaped data:", data_batches.shape)

# A
batch = data_batches[0]
print(batch.shape)

batch_transposed = batch.T
print(batch_transposed.shape)

```

```python
# Q-15:You have daily sensor data with shape(365,24)->365 days 24 hourly readings each day.

# (a)Flatten it into (8760,).

# (b)Add an extra dimension to make (365,24,1) for ML input.

# (c)Use sequeeze to bring it back to (365,24).

sensor_data = np.random.rand(365, 24)

# a
flattened = sensor_data.flatten()
print("Flatten shape:", flattened.shape)

# b
sensor_expanded = sensor_data[..., np.newaxis]
print("Expanded shape:", sensor_expanded.shape)

# c
sensor_squeezed = sensor_expanded.squeeze()
print("Squeezed shape:", sensor_squeezed.shape)
```

```python
# Q-16: You have stock prices with shape(30,5)->30 days, 5 companies.

# (a)Extract one company's stock price as a 1D array(30,).

# (b)Convert the same into a column vector (30,1).

# (c)Transpose the dataset to (5,30)->stocks*days.

stock_prices= np.random.rand(30, 5)
company_1d = stock_prices[:, 0]
company_col = company_1d.reshape(-1, 1)
stock_transposed = stock_prices.T

print("Company 1d shape:", company_1d.shape)
print("Company column shape:", company_col.shape)
print("Transpose shape:", stock_transposed.shape)
print(stock_transposed)
print(stock_prices.shape)
```

```python
# Q-18:You have 200 senetences,each represented by 50 words,each word embedding of 100 dimensions.

# ->shape(200,50,100)

# (a)Flatten each sentence into (200,5000) for a bag-of-word style model.

# (b)Transpose it to (200,100,50) so embeddings come before word positions.

# (c)Extract one sentence and reshape into (50,100)again.

senetences = np.random.rand(200, 50, 100)

# a
flattened_sen = senetences.reshape(200, 5000)
print("Flattened shape:", flattened_sen.shape)

# b
transposed_sen = senetences.transpose(0, 2, 1)
print("Transposed shape:", transposed_sen.shape)

# c
one_senetences = senetences[0].reshape(50, 100)
print("One senetence shape:", one_senetences.shape)
print(one_senetences)
```
