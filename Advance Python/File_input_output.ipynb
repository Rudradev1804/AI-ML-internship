```python
# Q-4 Check if a file exists before trying to open it.
import os

filename = 'intro.txt'

if os.path.exists(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        print(f.read())
else:
    print(f"File '{filename}' does not exist.")

file.close()
```

```python
# Q-5 Append one new line to an existing file without deleting old content.

file = open("intro.txt", 'a')
file.write("\nI have also 6 months of experience in pyhton.")
file.close()
```

```python
# Q-6 Write a program that reads a file and converts all text to uppercase.
try:
    with open('new.txt', 'r') as fin:
        text = fin.read()
except FileNotFoundError:
    with open(filename, 'r') as fin:
        text = fin.read()

upper_text = text.upper()
print(upper_text)
```
