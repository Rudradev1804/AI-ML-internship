```python
#Q-1. Filter out numbers greater than 5.

lst = [3, 7, 2, 9, 12, 17, 4]

filtered_lst = list(filter(lambda x: x >= 5, lst))
print(filtered_lst)
```

```python
# Q2. Given Filter only the odd numbers.
lst = [10, 21, 32, 43, 54, 65]

odd_num=list(filter(lambda x: x%2 !=0, lst))
print(odd_num)
```

```python
# Q3. Given Filter names that have more than 4 letters.
names = ["Sumit", "Ranu", "Raj", "Aman", "Sita"]

large_name=list(filter(lambda x: len(x)>4, names))
print(large_name)
```

```python
# Q4. Given Filter words that start with 'b'.
words = ["apple", "cat", "banana", "dog"]

word_=list(filter(lambda x: x.startswith('b'), words))
print(word_)
```

```python
# Q5. Given Use map to create a list of each number multiplied by 10.
lst = [1, 2, 3, 4, 5]

multi_lst=list(map(lambda x: x*10, lst))
print(multi_lst)
```

```python
# Q6. Given Increase each price by 15% using map.
prices = [100, 200, 300]

increased_prices=list(map(lambda x: x*1.15, prices))
print(increased_prices)
```

```python
# Q7. Given Convert each word to uppercase using map.
words = ["hello", "python", "world"]

uppercase_words=list(map(lambda x: x.upper(), words))
print(uppercase_words)
```

```python
# Q8. Given Convert each number to its string using map.
lst = [2, 4, 6, 8]

string_lst=list(map(lambda x: str(x), lst))
print(string_lst)
```

```python
# Q9. Given Use reduce to find the product of all numbers.
import functools
lst = [1, 2, 3, 4, 5]

product=functools.reduce(lambda x, y: x * y, lst)
print(product)
```

```python
# Q10. Given Use reduce to find the largest number.
import functools
lst = [100, 20, 50]

largest=functools.reduce(lambda x, y: x if x > y else y, lst)
print(largest)
```

```python
# Q11. Given Use reduce to join words into a sentence ‚ûù "I love Python"
import functools
words = ["I", "love", "Python"]

sentence=functools.reduce(lambda x, y: x+" "+y, words)
print(sentence)
```

```python
# Q12. Given Use reduce to find the difference like this: ((((5 - 10) - 15) - 20))

import functools
lst = [5, 10, 15, 20]

difference=functools.reduce(lambda x, y: x - y, lst)
print(difference)
```
