```python
# Filter out numbers greater than 5.

lst = [3, 7, 2, 9, 12, 17, 4]

filtered_lst = list(filter(lambda x: x >= 5, lst))
print(filtered_lst)
```

```python
# Q2. Given Filter only the odd numbers.
lst = [10, 21, 32, 43, 54, 65]

odd_num=list(filter(lambda x: x%2 !=0, lst))
print(odd_num)
```

```python
# Q3. Given Filter names that have more than 4 letters.
names = ["Sumit", "Ranu", "Raj", "Aman", "Sita"]

large_name=list(filter(lambda x: len(x)>4, names))
print(large_name)
```

```python
# Q4. Given Filter words that start with 'b'.
words = ["apple", "cat", "banana", "dog"]

word_=list(filter(lambda x: x.startswith('b'), words))
print(word_)
```

```python
# Q5. Given Use map to create a list of each number multiplied by 10.
lst = [1, 2, 3, 4, 5]

multi_lst=list(map(lambda x: x*10, lst))
print(multi_lst)
```
