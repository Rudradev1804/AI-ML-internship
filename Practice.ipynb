```python
# 4. Take no.of days from user and convert it into days, months, year
numberOfDays = int(input("Enter number of days: "))
year = 365
month = 30
years = numberOfDays // year
remaining_days_after_years = numberOfDays % year
months = remaining_days_after_years // month
days = remaining_days_after_years % month

print(f"Years = {years}")
print(f"Months = {months}")
print(f"Days = {days}")
```

```python
# 4. Take no.of days from user and convert it into days,months,year

days = int(input("Enter number of days: "))
years = days // 365
remaining_days = days % 365
months = remaining_days // 30
days_left = remaining_days % 30

print(f"{days} days = {years} years") 
print(f"{months} months") 
print(f"{days_left} days")
```

```python
# 5. Take no.of seconds from user and convert it into seconds,minutes,hours

total_seconds = int(input("Enter total number of seconds: "))
seconds_in_minute = 60
seconds_in_hour = 3600
seconds_in_day = 86400
days = total_seconds // seconds_in_day
remaining_seconds_after_days = total_seconds % seconds_in_day
hours = remaining_seconds_after_days // seconds_in_hour
remaining_seconds_after_hours = remaining_seconds_after_days % seconds_in_hour
minutes = remaining_seconds_after_hours // seconds_in_minute
seconds = remaining_seconds_after_hours % seconds_in_minute

print(f"{total_seconds} seconds = {days} days") 
print(f"{seconds} seconds")
print(f"{minutes} minutes") 
print(f"{hours} hours") 
```
