```python
# Q-1: Create class Calculator with a method add() that can:

# ->add two integers.

# ->add three integers.

# ->add two strings(concatation).

# Here you need to use method overloading(by default arguments in python).

class Calculator:
    def add(self, *args):
        if len(args) == 2:
            return args[0] + args[1]
        elif len(args) == 3:
            return args[0] + args[1] +args[2]
        else:
            return "Invalid numbers of arguments"
        
calc = Calculator()
print(calc.add(2, 3))
print(calc.add(1, 2, 3))
print(calc.add("Hello", "World"))
```

```python
# Q-2: Create a base class Employee with method get_role().

# ->Subclass Developer should override get_role() to return "Writes Code".

# ->Subclass Designer should override get_role() to return "Creates Designs".

# ==>Here You Use method overriding.

class Employee:
    def get_role(self):
        return "Employee Role"
    
class Developer(Employee):
    def get_role(self):
        return "Writes Code"
    
class Designer(Developer):
    def get_role(self):
        return "Creates Designs"
    
dev = Developer()
des = Designer()
print(dev.get_role())
print(des.get_role())
```

```python
# Q-3: Create a class Vector with attributes x and y.

# ->Overload thr + operator so that two vectors can be added.

# ->Example: v1+v2 should return a new Vector with summed coordinates.

# =>here you must use operator overloading(__add__).

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(v3)
```

```python
# Q-4: Create two classes PDFReader and DocReader, each with a method read().

# ->Write a function open_file(reader) that calls reader.read(),regardless of the object type.

# =>here you use duck typing.

from anyio import open_file


class PDFReader:
    def read(self):
        return "Reading PDF files"
    
class DocReader:
    def read(self):
        return "Reading DOC files"

    def open_file(reader):
        return reader.read()
    
reader = PDFReader()
print(open_file(reader))
reader = DocReader()
print(open_file(reader))
```

```python
# Q-5: Create an abstract class Aplliance with abstract method turn_on().

# ->subclass Fan,Light and Heater must implement turn_on() differently.

# ->Write a function start_all(Appliance) that loops over a list of appliances and calls turn_on().

# =>here you must use Abstract Class(ABC)+Overriding.

from abc import ABC, abstractmethod

class Application(ABC):
    @abstractmethod
    def turn_on(self):
        pass

class Fan(Application):
    def turn_on(self):
        return "Fan is noe ON"

class Light(Application):
    def turn_on(self):
        return "Light is now ON"
    
class Heater(Application):
    def turn_on(self):
        return "Heater is now ON"
    
def start_all(application):
    for app in application:
        print(app.turn_on())

fan = Fan()
light = Light()
heat = Heater()
apps = [fan, light, heat]
start_all(apps)
```

```python
# Q-6: Create a class hierarchy for a Payment System.

# ->Base Class: Payment(abstract method make_payment(amount)).

# ->Subclass: CreditCard,Paypal,UPI-->override make_payment().

# ->Overloading: Add a method transaction() that can take 1 or 2 arguments(amount only, or amount+currency).

# ->Operator Overloading: Overloading+so two payments can be combined (eg: PayPal+UPI=combined balance).

# ->Duck Typing: Create a function process_payment(payment_method) that works with any class that has make_payment().
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def make_payment(self, amount):
        pass
    def transaction(self, amount, currency=None):
        if currency is None:
            return f"Processsing payment of {amount}"
        else:
            return f"Processing payment of {amount} in {currency}"
        
class CreditCard(Payment):
    def make_payment(self, amount):
        return f"CreditCard Payment of {amount} made"
    
class Paypal(Payment):
    def __init__(self, balance):
        self.balance = balance
    def make_payment(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"Payment of {amount} made via PayPal"
        else:
            return "Insufficient balance"
    def __add__(self, other):
        return Paypal(self.balance + other.balance)
    
class UPI(Payment):
    def __init__(self, balance):
        self.balance = balance
    def make_payment(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"Payment of {amount} made via UPI"
        else:
            return "Insufficient balance"
        
    def __add__(self, other):
        return UPI(self.balance + other.balance)
    
    def process_payment(payment_method, amount):
        return payment_method.make_payment(amount)
    
cc = CreditCard()
print(cc.make_payment(100))
pay = Paypal(500)
upi = UPI(300)
print(pay.make_payment(200))
print(upi.make_payment(400))

combined_payment = pay + upi
print(f"Combined Balance: {combined_payment.balance}")
print(cc.transaction(150))
```
