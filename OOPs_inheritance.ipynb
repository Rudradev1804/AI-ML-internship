```python
# Q-1: Create class Calculator with a method add() that can:

# ->add two integers.

# ->add three integers.

# ->add two strings(concatation).

# Here you need to use method overloading(by default arguments in python).

class Calculator:
    def add(self, *args):
        if len(args) == 2:
            return args[0] + args[1]
        elif len(args) == 3:
            return args[0] + args[1] +args[2]
        else:
            return "Invalid numbers of arguments"
        
calc = Calculator()
print(calc.add(2, 3))
print(calc.add(1, 2, 3))
print(calc.add("Hello", "World"))
```

```python
# Q-2: Create a base class Employee with method get_role().

# ->Subclass Developer should override get_role() to return "Writes Code".

# ->Subclass Designer should override get_role() to return "Creates Designs".

# ==>Here You Use method overriding.

class Employee:
    def get_role(self):
        return "Employee Role"
    
class Developer(Employee):
    def get_role(self):
        return "Writes Code"
    
class Designer(Developer):
    def get_role(self):
        return "Creates Designs"
    
dev = Developer()
des = Designer()
print(dev.get_role())
print(des.get_role())
```

```python
# Q-3: Create a class Vector with attributes x and y.

# ->Overload thr + operator so that two vectors can be added.

# ->Example: v1+v2 should return a new Vector with summed coordinates.

# =>here you must use operator overloading(__add__).

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(v3)
```
