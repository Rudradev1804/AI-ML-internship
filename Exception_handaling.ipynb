```python
# Q-1 Write a program that handles division by zero error.

try:
    print(10/0)
except ZeroDivisionError:
    print("Number is not divisable by zero!")
```

```python
# Q-2 Take user input and handle ValueError if input is not a number.

try:
    num = int(input("Enter a number: "))
    print(num/0)
except ValueError:
    print("Please enter a number.")
```

```python
# Q-3 Handle both ZeroDivisionError and ValueError together.

try:
    num = int(input("Enter a number: "))
    print(num/0)
except ValueError:
    print("Please enter a number.")
except ZeroDivisionError:
    print("Number is not divisable by zero")
```

```python
# Q-4 Try reading a non-existing file and handle the FileNotFoundError.

try:
    with open('hello.txt', 'r') as f:
        data = f.read()
        print(data)
except FileNotFoundError:
    print("This file is not exist")
```

```python
# Q-5 Add an else block that prints a success message if no error occurs.

try:
    num = int(input("Enter a number: "))
    print(num/1)
except ZeroDivisionError:
    print("Please enter a number.")

else:
    print("This is else block message")
```

```python
# Q-6 Write a program to handle multiple specific exceptions in one try.

try:
    with open('hello.txt', 'r') as f:
        data = f.read()
        print(data)
        
    num = int(input("Enter a number: "))
    print(num/0)

except FileNotFoundError:
    print("This file is not exist")
except ValueError:
    print("Please enter a number.")
except ZeroDivisionError:
    print("Number is not divisable by zero")
```

```python
# Q-7 Create a program that always prints "Program ended" using finally.

try:
    num = int(input("Enter a number: "))
    print(num/0)
except ZeroDivisionError:
    print("Number is not divisable by zero.")

finally:
    print("This message was print every time")
```
