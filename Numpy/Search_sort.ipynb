```python
# Dataset 1 â€” Students Marks

# marks = np.array([35, 50, 65, 80, 40, 90, 55, 70])

# Q11: Find indexes of students who passed (marks â‰¥ 50).

# Q12: Replace all marks below 50 with 0 using np.where().

# Q13: Find where a new score 75 should be inserted to keep marks sorted.

# Q14: Get all scores greater than average marks.
import numpy as np

marks = np.array([35, 50, 65, 80, 40, 90, 55, 70])
avg = np.where(marks > 50)
print(avg)

rpl = np.where(marks<50, 0, marks)
print(rpl)

new = np.searchsorted(marks, 75)
print(new)

great = np.mean(marks)
greater = marks[marks>great]
print(great)
print(greater)
```

```python
# ðŸ§© Dataset 5 â€” Daily Steps (Fitness Tracker)

# steps = np.array([4000, 7500, 12000, 3000, 10000, 6000, 8000])

# Q27: Find indexes of days with steps > 8000.

# Q28: Replace steps below 5000 with 'Low Activity' and others with 'Active'.

# Q29: Get all days where steps are below the weekly average.

# Q30: Check using np.any() if any day has 0 steps (missed logging).

steps = np.array([4000, 7500, 12000, 3000, 10000, 6000, 8000])

activate = np.where(steps > 8000)
print(activate)

replace = np.where(steps < 5000, 'Activity', 'Activate')
print(replace)

aver = steps.mean()
print(aver)

below_avg_days = np.where(steps < avg)[0]
below_avg_steps = steps[below_avg_days]
print(below_avg_days)
print(below_avg_steps)

zero = np.any(steps == 0)
print(zero)
```
