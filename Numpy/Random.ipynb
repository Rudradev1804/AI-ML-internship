```python
# Q1.Generate an array of 5 random float numbers between 0 and 1 using np.random.rand().

import numpy as np

rnum = np.random.rand(5)
print(rnum)
```

```python
# Q2.Create a 2D random array (shape = 3x4) with random numbers between 0 and 1.

arr2d = np.random.rand(3, 4)
print(arr2d)
```

```python
# Q3.Generate 10 random integers between 50 and 100 using np.random.randint().
rnum = np.random.randint(50, 101, size=10)
print(rnum)
```

```python
# Q4.Create a 3x3 matrix of random integers between 1 and 9.

matrix = np.random.randint(1, 10, size=(3, 3))
print(matrix)
```

```python
# Q5.Generate 5 random numbers from a normal distribution using np.random.randn().

num = np.random.randn(5)
print(num)
```

```python
# Q6.Use np.random.choice() to select 3 random elements from the array:

arr = np.array([10, 20, 30, 40, 50])
select = np.random.choice(arr, size=3, replace=False)
print(select)
```

```python
# Q7.Use np.random.choice() to pick 5 elements with replacement from:

arr = np.array([1, 2, 3, 4, 5])
select = np.random.choice(arr, size=5, replace=True)
print(select)
```

```python
# Q8.Shuffle this array using np.random.shuffle():

arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)
```

```python
# Q10.

# Generate a 4x4 matrix of random integers between 10 and 99 and find:

# Maximum value

# Minimum value

# Mean

matrix = np.random.randint(10, 100, size=(4, 4))
max_value = np.max(matrix)
min_value = np.min(matrix)
mean_value = np.mean(matrix)
print("Matrix:\n", matrix)
print("Maximum value:", max_value)
print("Minimum value:", min_value)
print("Mean value:", mean_value)
```
