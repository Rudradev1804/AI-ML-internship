```python
# Q-1:Create two 1D arrays [1,2,3] and [4,5,6].

# ->Join them using np.concatenate.

# ->Join them using np.hstack.

# ->What is the diffrence in output?
import numpy as np

arr1d = [1, 2, 3]
arr1d1 = [4, 5, 6]

result_concat = np.concatenate((arr1d, arr1d1))
result_hstack = np.hstack((arr1d, arr1d1))
print("Using np.concatenate:", result_concat)
print("Using np.hstack:", result_hstack)
```

```python
# Q-2:Given arrays:

# a=np.array([[1,2],[3,4]])
# b=np.array([[5,6],[7,8]])

# ->Join them row-wise using concatenate.

# ->Join them column-wise using concatenate.

a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])

print("Row wise:", np.concatenate((a, b), axis=0))
print(np.concatenate((a, b), axis=1))
```

```python
# Q-3:Create two 1D arrays and join them into a 2D column matrix using np.column_stack.

arr = [1, 2, 3]
arr1d = [4, 5, 6]
result = np.column_stack((arr, arr1d))
print("Using np.column:", result)
```

```python
# Q-4:Create arrays:

# a=np.array([1,2,3])
# b=np.array([4,5,6])

# ->Stack them using np.stack with axis=0

# ->Stack them using np.stack with axis=1

# ->Explain the diffrence in shape.

a=np.array([1,2,3])
b=np.array([4,5,6])

print(np.stack((a, b), axis=0))
print(np.stack((a, b), axis=1))
```

```python
# Q-5:Create a (2,2) array filled with ones and another (2,2) array filled with zeros.

# ->Stack them along depth using np.dstack.

# ->Check the shape and explain.

ones = np.ones((2,2))
zeros = np.zeros((2,2))

result = np.dstack((ones, zeros))
print(result)
print(result.shape)
```

```python
# Q-6:Create three arrays [10,20,30],[40,50,60],[0,80,90].

# ->Join them using np.vstack.

# ->Join them using np.row_stack.

# ->Are the results same?

arr1d = [10,20,30]
arr1d1 = [40,50,60]
arr1d2 = [0,80,90]
result_vstack = np.vstack((arr1d, arr1d1, arr1d2))
result_row_stack = np.hstack((arr1d, arr1d1, arr1d2))

print(result_vstack)
print(result_row_stack)
```

```python
# Q-8:You have 3 grayscale images of size (2,2):

# R=np.array([[255,0],[0,255]]) #red channel
# G=np.array([[0,255],[255,0]]) #green channel
# B=np.array([[100,100],[100,100]]) #blue channel


# ->Combine them into a single RGB image of shape(2,2,3) using np.dstack.

R=np.array([[255,0],[0,255]]) #red channel
G=np.array([[0,255],[255,0]]) #green channel
B=np.array([[100,100],[100,100]]) #blue channel

rgb_image = np.dstack((R,G,B))
print(rgb_image)
print(rgb_image.shape)
```

```python
# Q-1: You have a dataset of employee salaries(in thousands):

# salaries=np.array([
#     [101,45000,3],
#     [102,52000,4],
#     [103,58000,5],
#     [104,61000,6],
#     [105,70000,8],
#     [106,75000,9]
# ])

# Each row=[Employee_ID,Salary,Experience]

# Task: Split this dataset vertically into 3 columns(so you get ID,Salary and Experience seperately).

salaries=np.array([
    [101,45000,3],
    [102,52000,4],
    [103,58000,5],
    [104,61000,6],
    [105,70000,8],
    [106,75000,9]
])

emp_id, salary, experience = np.hsplit(salaries, 3)
print("Employee IDs:\n", emp_id)
print("Salaries:\n", salary)
print("Experience:\n", experience)
```

```python
# Q-2: Monthly sales(in thousands) of two branches combined:

# sales=np.array([
#     [12,15,18,20,23,25],
#     [14,17,20,22,26,30]
# ])

# Task: Split the array horizontally into 2 equal halves.

# ->First half=Jan-Mar

# ->Second half=Apr-Jun

sales=np.array([
    [12,15,18,20,23,25],
    [14,17,20,22,26,30]
])
jan_mar, apr_jun = np.hsplit(sales, 2)
print("First half (Jan-Mar):\n", jan_mar)
print("Second half (Apr-Jun):\n", apr_jun)
```

```python
# Q-3: You have quarterly revenue data for 3 years:

# revenue=np.array([
#     [100,120,130,150],
#     [200,230,250,270],
#     [300,320,350,380]
# ])

# Task: Split this into 4 parts- each part represents one quarter (Q1,Q2,Q3,Q4).

revenue=np.array([
    [100,120,130,150],
    [200,230,250,270],
    [300,320,350,380]
])
q1, q2, q3, q4 = np.hsplit(revenue, 4)
print("Q1:\n", q1)
print("Q2:\n", q2)
print("Q3:\n", q3)
print("Q4:\n", q4)
```

```python
# Q-4: marks=np.array([
#     [78,82,89,91],
#     [85,87,90,93],
#     [75,80,83,88],
#     [92,95,97,99]
# ])

# Each row=student
# Each column=subject

# Task: Split students into 2 equal groups(top half,bottom half).

marks=np.array([
    [78,82,89,91],
    [85,87,90,93],
    [75,80,83,88],
    [92,95,97,99]
])
top_half, bottom_half = np.vsplit(marks, 2)
print("Top half (first 2 students):\n", top_half)
print("Bottom half (last 2 students):\n", bottom_half)
```

```python
# Q-5: data=np.array([
#     [[10,20],[30,40]],
#     [[50,60],[70,80]]
# ])

# Here Each depth represents 2 diffrent months(Jan and Feb).

# Task: Split this array along depth(axis=2)->so that you get 2 arrays,one for each month.

data=np.array([
    [[10,20],[30,40]],
    [[50,60],[70,80]]
])
jan, feb = np.dsplit(data, 2)
print("January data:\n", jan)
print("February data:\n", feb)
```

```python
# Q-6: transactions=np.array([
#     [2000,2500,3000,3500,4000,4500],
#     [1800,2200,2700,3200,3700,4200]
# ])

# Each row->Account holder
# Each column->Month(Jan to Jun)

# Task-1: Split into 3 parts(each part=2 months).

# Task-2: Find total amount of each split part using np.sum()

transactions=np.array([
    [2000,2500,3000,3500,4000,4500],
    [1800,2200,2700,3200,3700,4200]
])
jan_feb, mar_apr, may_jun = np.hsplit(transactions, 3)
print("Jan-Feb:\n", jan_feb)
print("Mar-Apr:\n", mar_apr)
print("May-Jun:\n", may_jun)
print("Total for Jan-Feb:", np.sum(jan_feb))
print("Total for Mar-Apr:", np.sum(mar_apr))
print("Total for May-Jun:", np.sum(may_jun))
```

```python
# Q-7: feedback=np.array([
#     [4,5,3,4,2,5],
#     [5,4,4,5,3,4],
#     [3,4,2,3,4,5]
# ])

# Each row=Customer
# Each column=Rating given over 6 months

# Task: Split ratings into first 3 months and last 3 months.

# Then calculate average rating in both parts for each customer.

feedback=np.array([
    [4,5,3,4,2,5],
    [5,4,4,5,3,4],
    [3,4,2,3,4,5]
])

first_half, last_half = np.hsplit(feedback, 2)
print(first_half)
print(last_half)
print(np.mean(first_half, axis=1))
print(np.mean(last_half, axis=1))
print(np.mean(first_half))
print(np.mean(last_half))
```
