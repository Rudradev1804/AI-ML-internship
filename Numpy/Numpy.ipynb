```python
# Q-1: Create a 1D array with values [5,10,15,20,25].

# ->Print the array.

# ->Print its dimension and shape.

import numpy as np

arr1d = np.array([5,10,15,20,25])
print(arr1d)
print(arr1d.ndim)
print(arr1d.shape)
```

```python
# Q-2: Create a 1D array of numbers from 0 to 9.

# ->Verify that its 1D.

# ->Print its shape.

import numpy as np

arr1d = np.array(range(10))
print(arr1d)
print(arr1d.ndim)
print(arr1d.shape)
```

```python
# Q-6 Create a 3D array with random integers between 1 and 20. Shape should be(2,2,3).

# ->Print the array.

# ->Print its dimension and shape.

arr3d = np.random.randint(1,20, size = (2,2,3))
print(arr3d)
print(arr3d.ndim)
print(arr3d.shape)
```

```python
# Q-8: Given array b=[[1,2,3],[4,5,6]],create zeros_like(b) and ones_like(b). 

b=np.array([[1,2,3],[4,5,6]])
print("Zeros like b\n",np.zeros_like(b))
print("Ones like b\n",np.ones_like(b))
```
