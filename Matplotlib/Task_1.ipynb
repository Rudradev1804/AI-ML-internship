```python
# Create x = [1,2,3,4,5] (tests)
# Create y = [45, 50, 60, 70, 80] (marks)
from matplotlib import legend
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45, 50, 60, 70, 80]
plt.plot(x,y, marker='o', linestyle='dashed', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('line graph')
plt.show()
```

```python
# Create temperature values for 7 days
# Rotate x-axis labels by 30 degrees
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
tem = [22, 24, 19, 23, 25, 21, 20]
plt.plot(days, tem, marker='s', linestyle='dashed', color='red')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=30)
plt.title('Temperature over a Week')
plt.show()
```

```python
# Generate 20 random values using np.random.randint
import numpy as np
ran = np.random.randint(1, 100, 20)
print(ran)
plt.hist(ran, linestyle='dotted', color='green', edgecolor='black')
plt.xlabel('Values Range')
plt.ylabel('Frequency')
plt.title('Histogram of Random Values')
plt.grid(True)
plt.show()
```

```python
x = np.random.randint(1, 50, size=15)
y = np.random.randint(1, 100, size=15)

plt.figure(figsize=(8,6))
plt.plot(x, y, marker='^', markersize=8, linestyle='-', color='blue')
plt.xlabel("X values (1 to 50)")
plt.ylabel("Y values (1 to 100)")
plt.title("Random Integer Plot")
plt.grid(True)
plt.show()
```
