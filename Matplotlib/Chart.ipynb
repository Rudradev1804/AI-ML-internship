```python
# Create a pie chart showing favorite fruits of students:
import matplotlib.pyplot as plt

fruits = ['Apples', 'Bananas', 'Cherries', 'Mangoes']
value = [25, 30, 20, 25]
colors = ['yellow', 'cyan', 'limegreen', 'greenyellow']
legends = ('Apples: 25%', 'Bananas: 30%', 'Cherries: 20%', 'Mangoes: 25%')
plt.pie(value, labels=fruits, autopct='%1.1f%%', startangle=0, colors=colors)
plt.title("Favoutite fruits of students")
plt.legend(legends, loc ='upper left')
plt.show()
```

```python
# Create a pie chart of monthly expenses:

month_exp = ['Rent', 'Food', 'Travel', 'Savings']
expence = [12000, 5000, 55000, 30000]
colors = ['lightcoral', 'gold', 'lightskyblue', 'lightgreen']
legends = ('Rent', 'Food', 'Travel', 'Savings')
plt.pie(expence, labels=month_exp, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Monthly Expenses")
plt.legend(legends, loc = 'lower right')
plt.show()
```

```python
# Create an area chart showing hours vs marks:
hours = [1, 2, 3, 4, 5]
marks = [10, 20, 40, 60, 80]
plt.fill_between(hours, marks, color='skyblue', alpha=0.5)
plt.plot(hours, marks, color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Hours vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.show()
```
