```python
from operator import le
import matplotlib.pyplot as plt

student = ['A', 'B', 'C', 'D', 'E']
marks = [85, 90, 78, 92, 88]
colors = ['blue', 'orange', 'green', 'red', 'purple']
legend = ['A', 'B', 'C', 'D', 'E']
plt.barh(student, marks, color = colors)
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Student Marks')
plt.legend(legend)
plt.show()
```

```python
expence = ['Rent', 'Food', 'Transport', 'Savings']
amount = [17000, 5000, 3500, 10000]
colors=['red','skyblue','pink','green']
plt.pie(amount, labels=expence, autopct='%1.1f%%', startangle=90, explode=(0, 0.1, 0.1, 0.1), shadow=True, colors=colors)
plt.title('Monthly expence')
plt.show()
```
