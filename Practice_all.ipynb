```python
                                                            # List
# Q-1 Write a program to find the largest number in a list.

lst = [1,2,3,5,6,7,8,4,9]
largest = max(lst)
print(f"The largest number in the list is: {largest}")
```

```python
# Q-2 Remove duplicates from a list without using set().
lst = [1,2,3,4,5,6,6,3,7,8,4,5]
duplicate = []
for items in lst:
    if items not in duplicate:
        duplicate.append(items)

print(f"Original list : {lst}")
print(f"Wothout dulicates : {duplicate}")
```

```python
# Q-3 Count how many times each element appears in a list.
lst = [1,2,3,4,5,6,6,3,7,8,4,5]

result = {i: lst.count(i) for i in set(lst)}
print(result)
```

```python
# Q-4 Reverse a list without using the reverse() function.
lis = [1,2,3,4,5,6,7,8,9]

reversed_list = lis[::-1]
print(reversed_list)
```

```python
# Q-5 Find the second largest number in a list.
lst = [1, 2, 3, 5, 6, 7, 8, 4, 9]
unique_lst = list(set(lst))
unique_lst.sort(reverse=True)

if len(unique_lst) >=2:
    second_largest = unique_lst[1]
    print(f"Second largest number is : {second_largest}")
else:
    print("There is ni list")
```

```python
# Q-6 Merge two lists and remove duplicates.
lst_1 = [1, 2, 3, 5, 6, 7, 8, 4, 9, 6, 7 ,3]
lst_2 = [1,2,3,4,5,6,6,3,7,8,4,5]

lst_3 = list(set(lst_1 + lst_2))
print(f"Merged list without duplicates: {lst_3}")
```

```python
                                                            # SET

# Q-1 Find common elements between two sets.
s_1 = (1,2,3,4,5,6,7)
s_2 = (2,3,5,6,8,9,7)

common = set(s_1) & set(s_2)
print(common)
```

```python
# Q-2 Check if one set is a subset of another.
s_1 = set(s_1)
s_2 = set(s_2)

if s_1.issubset(s_2):
    print(f"{s_1} is a subset of {s_2}")
else:
    print(f"{s_1} is not a subset of {s_2}")

if s_2.issubset(s_1):
    print(f"{s_2} is a subset of {s_1}")
else:
    print(f"{s_2} is not a subset of {s_1}")
```
