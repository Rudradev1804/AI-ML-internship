```python
# Q-1 Print today’s date using date.today().
from datetime import timedelta
import datetime

date = datetime.datetime.today()
print(date)
```

```python
# Q-2 Print current time using datetime.now().

date_time = datetime.datetime.now()
print(date_time)
```

```python
# Q-3 Extract year, month, and day from today’s date.

today_date = datetime.datetime.now()
print(today_date.strftime("%d"))
print(today_date.strftime("%m"))
print(today_date.strftime("%Y"))
```

```python
# Q-4 Create a date object for your birthday.

birthday = datetime.date(2004, 9, 18)
print(birthday)
```

```python
# Q-5 Print date after 30 days from today.

today_day = datetime.date.today()
print("Today's date is :", today_day)

next_day = today_day + datetime.timedelta(days = 30)

print("Date after 30 days from today is :", next_day)
```

```python
# Q-6 Subtract 15 days from today’s date.

before = today_day - datetime.timedelta(days=15)
print("Before 15 days the date is :", before)
```

```python
# Q-7 Find how many days left for New Year (Jan 1, 2026).

custome = datetime.date(2026, 1, 1)
difference = custome - today_day
print("Days left for new year :", difference.days)
```

```python
# Q-8 Convert "2025-05-15" string into a date object.
import datetime

string_date = "15-05-2025"
date_object = datetime.datetime.strptime(string_date, "%d-%m-%Y")
print(date_object)
```

```python
# Q-9 Print current date in format DD-MMM-YYYY.

current_date = datetime.datetime.now()
formate_date = current_date.strftime("%d-%B-%Y-%A-%p")
print("Current date in format DD-MMM-YYYY is :", formate_date)
```

```python
# Q-10 Print current time in 12-hour format with AM/PM.

current_datetime = datetime.datetime.now()
time_12hr = current_datetime.strftime("%I:%M %p")

print("current_datetime in 12 hours format with AM/PM :", time_12hr)
```

```python
# Q-11 Display current day name.

current_day = datetime.datetime.now()
name_of_day = current_day.strftime("%A")
print("Current day name is :", name_of_day)
```

```python
# Q-12 Add 2 hours and 30 minutes to current time.

current_time = datetime.datetime.now()
time = current_time + datetime.timedelta(hours=2, minutes=30)

print("Time after 2 Hours and 30 Minutes :", time.strftime("%H:%M"))
```

```python
# Q-13 Print only the month name of current date.

current_month = datetime.datetime.now()
month_name = current_month.strftime("%B")

print("Current month name is :", month_name)
```

```python
# Q-14 Format DateTime

now = datetime.datetime(2020, 2, 25)
formatted_date = now.strftime("%A %d %B %Y")

print("Formatted DateTime is :", formatted_date)
```

```python
# Q-15 Find Day of Week

date_time = datetime.datetime(2020, 2, 25)
day_of_week = date_time.strftime("%A")

print("Day of week is :", day_of_week)
```

```python
# Q-16 Add Week to Given Date

import datetime

given_date = datetime.datetime(2020, 3, 22, 10, 0, 0)
print("Given date is :", given_date)

new_date = given_date + datetime.timedelta(days=7, hours=12)
print("New date after adding one week is :", new_date)
```

```python
# Q-17 Print Current Time in Milliseconds

now = datetime.datetime.now()
print(now)

print(now.strftime("%I:%M:%S.%f %p"))
```

```python
# Q-18 Convert Datetime( into String

n = datetime.date(2020, 2, 25)
print(n)

string_date = n.strftime("%Y-%m-%d")
print("Converted string date is :", string_date)
```

```python
# Q-19 Calculate the date 4 months from the current date

current_date = datetime.datetime.now()
print("Current date is :", current_date)

after_four_months = current_date + datetime.timedelta(days=120)
print("Days after four months :", after_four_months)
```

```python
# Q-20 Calculate Days Between Two Dates

date_1 = datetime.datetime(2020, 2, 25)

date_2 = datetime.datetime(2020, 9, 17)

difference = date_1 - date_2
print("Days between two dates :", difference.days)
```
