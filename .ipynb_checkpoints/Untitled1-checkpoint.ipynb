```python
# 3. WAP to print string value of the number which is entered by user in int(0-9)
a = int(input("Enter a number between 0-9: "))
if a == 0:
    print("Zero")
elif a == 1:
    print("One")
elif a == 2:
    print("Two")
elif a == 3:
    print("Three")
elif a == 4:
    print("Four")
elif a == 5:
    print("Five")
elif a == 6:
    print("Six")
elif a == 7:
    print("Seven")
elif a == 8:
    print("Eight")
elif a == 9:
    print("Nine")
else:
    print("Number is not between 0-9")
```

```python
# 5. WAP to demonstrat use of calculator
n1 = eval(input("Enter the first number:"))
n2 = eval(input("Enter the second number:"))
op = input("Enter the operator (+,-,*,/):")
if op == '+':
    print(n1+n2)
elif op == '-':
    print(n1-n2)
elif op == '/':
    print(n1/n2)
elif op == '*':
    print(n1*n2)
else:
    print("Invalid operator", op)
```

```python
# 7. WAP to find grade by user entered percentage i.e (>90 A, >80 B, >70 C, >60 D)
per = eval(input("Enter your percentage:"))
if per > 90:
    print("Your grade is 'A'")
elif per > 80:
    print("Your grade is 'B'")
elif per > 70:
    print("Your grade is 'C'")
elif per > 60:
    print("Your grade is 'D'")
else:
    print("Your are fail 'F'")
```

```python
# 8. WAP to find type of triangle by taking value of 3 sides from user equilateral (all 3 sides are equal) isoceles (any 2 sides are equal) scalene(any sides are not equal)
a = eval(input("Enter side a:"))
b = eval(input("Enter side b:"))
c = eval(input("Enter side c:"))
if a==b and b==c:
    print("Equilateral triangle", a,b,c)
elif a==b or b==c or c==a:
    print("Isoceles triangle", a,b,c)
else:
    print("Scalene triangle")
```

```python
# 9. WAP to find if user entered year is leap year or not
y = eval(input("Enter a year: "))
if (y%4==0 and y%100!=0) or (y%400==0):
    print(y, "Is leap year")
else:
    print("Is not a leap year")
```

```python
# 1. WAP to find sum of 10 numbers entered by user
total = 0
for i in range(10):
    num = int(input("Enter the number: "))
    total += num

print("Final Sum is:", total)
```

```python
# 2. WAP to find average of 1st 10 numbers
s = 0
for i in range(1,11):
    s+=i
print("Average is :", s/10)
```

```python
# 3. WAP to take 3 numbers from user and check how many numbers between 1st entered number and 2nd entered numbers are divisible by 3rd number
a = int(input("Enter 1st number:"))
b = int(input("Enter 2nd number:"))
c = int(input("Enter 3rd number:"))
count = 0
for i in range(a, b+1):
    if i%c == 0:
        count+=1
print("Count of numbers between",a,"and",b,"divisable by",c,"is:",count)
```

```python
# 5. WAP to print multiplication table for a number entered by a user
n = int(input("Enter a number to print its multiplication table: "))
for i in range(1,11):
    print(n, 'x', i, '=', n*i)
```

```python
# 6. WAP to find factorial of a number entered by user
num = int(input("Enter the number to find its factorial: "))
fact = 1
for i in range(1, num+1):
    fact *=i
    print("Factorial of ", num, "is:", fact)
```

```python
# 8. WAP to find reversde of entered number
n1 = int(input("Enter a number to find its reverse: "))
rev = 0
while n1!=0:
    dig = n1%10
    rev = rev*10 + dig
    n1//=10
    print("Reverse is: ", rev)
```

```python
# 9. WAP to find sum of entered numbers
n2 = int(input("Enter a number to find its sum: "))
s = 0
while n2!=0:
    dig = n2%10
    s+=dig
    n2//=10
    print("Sum is:", s)
```

```python
# 10. WAP to find prime numbers from 1 to 100
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```

```python
# 11. WAP to calculate x to the power y
x = int(input("Enter base x: "))
y = int(input("Enter exponent y: "))

result = 1
for _ in range(y):
    result *= x

print(f"{x} to the power {y} is: {result}")
```
